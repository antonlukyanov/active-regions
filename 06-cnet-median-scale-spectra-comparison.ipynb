Критические сети строятся на фиксированном медианном масштабе для каждой области.

## Спектральный зазор

```python
dirs = [
    'AR2034H3996',
    'AR2035H4000',
    'AR2036H3999',
    'AR11158',
    'AR11166',
    'AR11261',
    'AR11283',
    'AR11402',
    'AR12173',
    'AR11542_wo_flares',
    'AR11654_wo_flares',
]

for imagesdir in dirs:
    imagesdir = 'data/images/' + imagesdir
    cn = CNet()
    cn.proc(imagesdir)
    cnf = CNet()
    scale = int(np.median(cn.stable_scales(5)))
    print('Scale: %d' % scale)
    cnf.proc(imagesdir, scale=scale)
    print()
    cnf.plot_spectral_gap()
```

## Алгебраическая связность

```python
for imagesdir in dirs:
    cn = CNet()
    cn.proc('images/' + imagesdir)
    cn.plot_algebraic_connectivity()
```
